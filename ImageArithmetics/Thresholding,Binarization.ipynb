```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def imshow(title="image",image = None, size =10):
    height, width = image.shape[0],image.shape[1]
    aspect_ratio = width/height
    plt.figure(figsize=(size*aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
    
```

```python
# Blurring using convolutions
image = cv2.imread(r"C:\Users\97798\Desktop\land.jpg",0)
imshow("Original Image",image)
```

```python
ret,thresh1 = cv2.threshold(image,200,255,cv2.THRESH_BINARY)
imshow("Threshold binary",thresh1)

ret,thresh2 = cv2.threshold(image,127,255,cv2.THRESH_BINARY_INV)
imshow("Threshold binary Inverse @127",thresh2)

ret,thresh3 = cv2.threshold(image,127,255,cv2.THRESH_TRUNC)
imshow("Threshold Trunc @127",thresh3)

ret,thresh4 = cv2.threshold(image,127,255,cv2.THRESH_TOZERO)
imshow("Threshold tozero @127",thresh4)

ret,thresh5 = cv2.threshold(image,127,255,cv2.THRESH_TOZERO_INV)
imshow("Threshold tozero Inv @127",thresh5)




```

```python
# Adaptive Thresholding
image = cv2.imread(r"C:\Users\97798\Desktop\land.jpg",0)
imshow("Original Image",image)

ret,thresh1 = cv2.threshold(image,127,255,cv2.THRESH_BINARY)
imshow("Threshold binary",thresh1)

thresh = cv2.adaptiveThreshold(image,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,3,5)
imshow('Adaptive thresolding',thresh)

_,th2 = cv2.threshold(image,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
imshow('OTSU thresholding',th2)

blur = cv2.GaussianBlur(image,(5,5),0)
_,th3 = cv2.threshold(blur,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
imshow("Gaussian Otsus Thresholding",th3)

```

```python
# Bilateral Blurr
bilateral = cv2.bilateralFilter(image,9,75,75)
imshow("Bilateral Blurring",bilateral)
```

```python
# Skimage Threshold Local
 
from skimage.filters import threshold_local

image = cv2.imread(r"C:\Users\97798\Desktop\land.jpg",1)
imshow("Original Image",image)

v= cv2.split(cv2.cvtColor(image,cv2.COLOR_BGR2HSV))[2]
T = threshold_local(v,25,offset=15,method = "gaussian")

thresh = (v>T).astype("uint8") *255
imshow("threshold_local",thresh)
```

```python
# Sharpening images
image = cv2.imread(r"C:\Users\97798\Desktop\land.jpg")
imshow("Original Image",image)

kernel_sharpening = np.array([[-1,-1,-1],
                              [-1,9,-1],
                              [-1,-1,-1],
                             ])

sharpened = cv2.filter2D(image,-1,kernel_sharpening)
imshow("Sharpened image",sharpened)
```
