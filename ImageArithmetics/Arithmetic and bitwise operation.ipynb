```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def imshow(title="image",image = None, size=10):
    w, h = image.shape[0],image.shape[1]
    asperct_ratio = w/h
    plt.figure(figsize=(size*asperct_ratio,size))
    plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
image = cv2.imread(r"C:\Users\97798\Desktop\land.jpg",0)
imshow('grayscale image',image)
print(image)
```

```python
M = np.ones(image.shape, dtype= "uint8") *100
print(M)
     
```

```python
# Increasing brightness
added = cv2.add(image,M)
imshow("Increased Brightness",added)
# Direct Addition
added2 = image+M
imshow("Increased Brightness with adding",added2)
```

```python
# Decreasing brightness
subtarct = cv2.subtract(image,M)
imshow("Decreased Brightness",subtarct)
# Direct Addition
added2 = image-M
imshow("Decreased Brightness with subtraction",added2)
```

```python
# Bitwise Opeartion

# Making a square

square = np.zeros((300,300),np.uint8)
cv2.rectangle(square,(50,50),(250,250),255,-2)
imshow("square",square)

# Making a ellipse 
ellipse = np.zeros((300,300),np.uint8)
cv2.ellipse(ellipse,(150,150),(150,150),30,0,100,255,-1)
imshow("ellipse",ellipse)
```

```python
# AND,OR

And = cv2.bitwise_and(square,ellipse)
imshow("AND",And)

OR = cv2.bitwise_or(square,ellipse)
imshow("OR",OR)

Xor = cv2.bitwise_xor(square,ellipse)
imshow("XOR",Xor)

Not = cv2.bitwise_not(square,ellipse)
imshow("Not",Not)
```
