```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

def imshow(title= "Image", image = None, size =10):
  w, h = image.shape[0], image.shape[1]
  aspect_ratio = w/h
  plt.figure(figsize=(size * aspect_ratio, size))
  plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
  plt.title(title)
  plt.show()



```

```python
image = cv2.imread('/aa.jpg', 0)
```

```python
imshow("Original", image)
```

```python
M = np.ones(image.shape, dtype="uint8") * 100
```

```python
print(M)
```

```python
added = cv2.add(image,M)

imshow("Increasing BrightNess", added)

added2 = image + M
imshow("Simple Numpy adding", added2)
```

```python
subtracted = cv2.subtract(image, M)
imshow("Subtracted", subtracted)
```

```python
square = np.zeros((300,300), np.uint8)
cv2.rectangle(square, (50,50),(250,250), 255, -2)
imshow("square", square)

ellipse = np.zeros((300,300), np.uint8)
cv2.ellipse(ellipse, (150,150), (150,150), 30,0,100,255, -1)
imshow("Ellipse", ellipse)
```

```python
And = cv2.bitwise_and(square, ellipse)
imshow("And Operation", And)

Or = cv2.bitwise_or(square, ellipse)
imshow("OR operations", Or)

```
