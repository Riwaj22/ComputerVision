```python
def imshow(title = "Image", image = None, size =10):
  w, h = image.shape[0], image.shape[1]
  aspect_ratio = w/h
  plt.figure(figsize = (size*aspect_ratio, size))
  plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
  plt.title(title)
  plt.show()
```

Drawing Polygons

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Create a black canvas
image = np.zeros((512, 512, 3), np.uint8)

# Define the vertices of the polygon
pts = np.array([[10, 10], [20, 20], [60, 60], [500, 80]], np.int32)
pts = pts.reshape((-1, 1, 2))

# Draw the polygon on the black canvas
cv2.polylines(image, [pts], isClosed=True, color=(0, 0, 255), thickness=3)

# Display the image using Matplotlib
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Black Canvas with Red Polygon")
plt.show()
```

Adding Text with cv2.putText
