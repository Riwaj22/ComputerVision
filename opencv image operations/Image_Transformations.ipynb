```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

def imshow(title= "Image", image = None, size =10):
  w, h = image.shape[0], image.shape[1]
  aspect_ratio = w/h
  plt.figure(figsize=(size * aspect_ratio, size))
  plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
  plt.title(title)
  plt.show()



```

Translations

```python
image = cv2.imread('/a.png')
```

```python
imshow("Original", image)
```

```python
height, width = image.shape[:2]

quarter_height, quarter_width = height/4, width/4

T = np.float32([[1,0, quarter_width],[0,1, quarter_height]

              ])

image_translation = cv2.warpAffine(image, T, (width, height))
imshow("Original Image", image)
imshow('Translated Image', image_translation)
```

Rotation

```python
rotation_matrix = cv2.getRotationMatrix2D((width/5, height/5), 90,1)
```

```python
rotated_image = cv2.warpAffine(image, rotation_matrix, (width, height))
imshow("Rotated Image", rotated_image)
```

```python
print(rotation_matrix)
```

```python
rotated_image = cv2.transpose(image)
imshow("Original Image", image)
imshow("Rotated Using Transpose", rotated_image)
```

```python
flipped = cv2.flip(image, 1)
imshow("Horizontal Flip", flipped)
```
