Downloading Images

```python
image = cv2.imread('/a.png')
```

```python
type(image)
```

Displaying Images

```python
from matplotlib import pyplot as plt

plt.imshow( cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
def imshow(title = "", image = None):

  plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
  plt.title(title)
  plt.show()
```

```python
imshow("Displaying an initial image", image)
```

Saving Images

```python
cv2.imwrite('output.png', image)
```
