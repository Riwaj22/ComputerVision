```python
import cv2
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
print(cv2.__version__)
```

```python
def imshow(title = "Image", image = None, size =10):
  w, h = image.shape[0], image.shape[1]
  aspect_ratio = w/h
  plt.figure(figsize = (size*aspect_ratio, size))
  plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
  plt.title(title)
  plt.show()
```

```python
B, G, R = cv2.split(image)
print(B.shape)
print(G.shape)
print(R.shape)
```

```python
imshow("Blue Channel Only", B
       )
```

```python
imshow("Green Channel Only", G
       )
```

```python
imshow("Red Channel Only", R
       )
```

```python
import numpy as np
zeros = np.zeros(image.shape[:2], dtype = np.uint8)

imshow("Red", cv2.merge([zeros, zeros, R]))
imshow("Green", cv2.merge([zeros, G, zeros]))
imshow("Blue", cv2.merge([B, zeros, zeros]))
```

```python
merged = cv2.merge([B,G,R])
imshow("merged", merged)
```

Lets Amplify the Blue Color


```python
merged = cv2.merge([B+100, G,R])
imshow("Blue Boost", merged)
```
